# SRT 예매 프로그램

```python
from selenium import webdriver

browser = webdriver.Chrome() # exe프로그램 | streamlit(로컬)
```

```python
# 로그인 페이지 이동
browser.get('https://etk.srail.kr/cmc/01/selectLoginForm.do?pageId=TK0701000000')
```

```python
from selenium.webdriver.common.by import By
import time

# 휴대폰 번호 버튼 클릭
browser.find_element(By.ID, 'srchDvCd3').click()

# 휴대폰 번호 / 비밀번호 입력
time.sleep(2)
browser.find_element(By.ID, 'srchDvNm03').send_keys('01012341234')
time.sleep(2)
browser.find_element(By.ID, 'hmpgPwdCphd03').send_keys('01012341234')

time.sleep(2)
# 확인 버튼 클릭
browser.find_element(By.XPATH, '//*[@id="login-form"]/fieldset/div[1]/div[1]/div[4]/div/div[2]/input').click()
```

```python
# //*[@id="result-form"]/fieldset/div[6]/table/tbody/tr[1]/td[6]/a # 20시
# //*[@id="result-form"]/fieldset/div[6]/table/tbody/tr[2]/td[6]/a # 20시 28분
# //*[@id="result-form"]/fieldset/div[6]/table/tbody/tr[3]/td[6]/a # 21시

num = 0
for i in range(30):
    num += 1
    text = browser.find_element(By.XPATH, '//*[@id="result-form"]/fieldset/div[6]/table/tbody/tr[1]/td[6]/a').text

    print(f'{num} 번째 실행 중입니다.')

    # browser.implicitly_wait(300)
    
    if text == '매진':
        browser.refresh()
    else:
        browser.find_element(By.XPATH, '//*[@id="result-form"]/fieldset/div[6]/table/tbody/tr[1]/td[6]/a').click()
```
